> TEXT =

> I heard the mission bell

> "This could be Heaven or this could be Hell"
> Then she lit up a candle and she showed me the way
> 

Average word length is the sum of all the lengths of the word tokens in the text, divided by the number of word tokens


```python
Text=""" 
 On a dark desert highway, cool wind in my hair
 Warm smell of colitas, rising up through the air
 Up ahead in the distance, I saw a shimmering light
 My head grew heavy and my sight grew dim
 I had to stop for the night
 There she stood in the doorway;
 I heard the mission bell
 And I was thinking to myself,
"This could be Heaven or this could be Hell"
 Then she lit up a candle and she showed me the way
 """

# Clean up the punctuation   
for char in '-.,;"':
    Text=Text.replace(char,' ')

# Replace '\n' with space    
for char in '\n':
    Text=Text.replace(char,' ')

# Create a list of words    
words = Text.split()

#Calculate word lenghts
word_len = [len(i) for i in words]

#Average word lengths
float(sum(word_len))/float(len(words))
```

(13) Write a program to make a list of all prime numbers less than 3000. A prime number can only be divided evenly by 1 and itself.

```python
# Create the list of numbers 1 to 3000
list_nos = range(1,3001)

# Create an empty list to collect all the non-prime numbers
non_prime =[]

# Initialize the list of primes to 2 as we already know that 2 is a prime
prime=[2] 

# Check for the divisibility of the number till the square root of the number. (According to Seive of Eratosthenes)
import math
for i in range(3,max(list_nos)):
    for j in range(2,int(math.ceil(math.sqrt(i)))+1):
            if i%j == 0:
                non_prime.append(i)
            else:
                prime.append(i)

# Calculate the final list of prime numbers                
final_prime = list(set(prime) - set(non_prime))


# Print allt he prime numbers
print final_prime
```

```python
# Create a set of numbers 1 to 3000
set_nos = set(range(1,3001))

# Create an empty set to collect the non_prime numbers
non_prime1 = set()

# Pre-populate the set with 2 as we know that 2 is a prime
prime1 = {2} 

# Check for the divisibility of the number till the square root of the number. (According to Seive of Eratosthenes)
for i in set_nos:
    for j in range(2,max(set_nos)):
        if i==j or i<j:
            continue
        else:
            if i%j == 0:
                non_prime1.add(i)
            else:
                prime1.add(i)


# Calculate the set of prime numbers                
final_prime1 = prime1 - non_prime1

# Print all the prime numbers
print final_prime1
      
```
